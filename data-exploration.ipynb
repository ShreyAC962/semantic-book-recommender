```python
import kagglehub

# Download the latest version of the dataset
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

# Print path to dataset
print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = plt.axes() # Create axes object
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax) # Plot heatmap of missing values

plt.xlabel("Columns") # Set x-axis label
plt.ylabel("Missing") # Set y-axis label
```

```python
import numpy as np
books["missing_description"] = np.where(books["description"].isna(),1,0)
books["age_of_books"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_books", "missing_description", "average_rating", "ratings_count"]
correlation_matrix = books[columns_of_interest].corr(method = "spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot = True, fmt = ".2f", cmap = "coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Spearman heatmap")
plt.show()
```

```python
books_missing = books[~(books["description"].isna()) &
    (~books["num_pages"].isna()) &
    (~books["published_year"].isna())&
    (~books["average_rating"].isna()) ]
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
books_missing_25_words
```

```python
books_missing_25_words["title_and_subtitle"] = np.where(
    books_missing_25_words["subtitle"].isna(),
    books_missing_25_words["title"],
    books_missing_25_words[["title", "subtitle"]]
        .fillna("")              # fill NaN with empty string
        .astype(str)
        .agg(": ".join, axis=1) # join title and subtitle
)
```
